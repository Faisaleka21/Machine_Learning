# Regresi Linear Sederhana

1. Fitur Import untuk Regresi

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

2. Kode" Selanjutnya

ini adalah kode pemanggilan data set.

```python
df = pd.read_csv('https://raw.githubusercontent.com/febbisena/DataMining/refs/heads/main/Ice%20Cream.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

df.isna untuk mengetahui junmlah data yang kosong

```python
df.isna().sum()
```

Dibawah hanya melihat sebaran data saja

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Temperature'], y=df['Revenue'],color='blue')

plt.title('Temperature vs Revenue')
plt.xlabel('Temperature')
plt.ylabel('Revenue')

plt.show()
```

```python
x = df[['Temperature']]
y = df ['Revenue']
```

```python
x
```

```python
y
```

dibawah untuk membagi data set menjadi data training

Ini adalah <b><i>conflusien matrix

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
print("x_train = ", len(x_train))
print("x_test = ", len(x_test))
print("y_train = ", len(y_train))
print("y_test = ", len(y_test))
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(15,10))
plt.scatter(y_test, y_pred)
plt.xlabel('actual')
plt.ylabel('predicted')
plt.title('Actual vs Predicted')
```

```python
plt.figure(figsize=(15,10))

plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.scatter(x_test, y_pred, color='red', label='Predicted')

plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Actual vs Predicted Revenue (Test Data)')
plt.legend()
plt.show()
```

Diatas yg merah = hasilnya, yang biru = dataset aslinya

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE = ',mse)
print('RMSE = ',rmse)
print('MAE = ',mae)
print('R2 Score = ',r2)
```

```python
intercept = model.intercept_
coefficient = model.coef_[0]

rumus_regresi = f"y = {intercept:.4f} + ({coefficient:.4f} * x)"
print(f'Intercept (a) = {intercept:.4f}')
print(f'Coefficient (b) = {coefficient:.4f}')
print(f'\nRumus Regresi : {rumus_regresi}')
```

```python
x_input = float(input("Masukkan nilai temperature: "))

y_pred_formula = intercept + (coefficient * x_input)
y_pred_model = model.predict(pd.DataFrame([[x_input]], columns = ['Temperature']))[0]

print("\nHasil berdasarkan model : ", round(y_pred_model, 4))
print("Hasil berdasarkan rumus : ", round(y_pred_formula,4))
```
